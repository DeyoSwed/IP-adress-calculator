```python
#Converts ip adress to binary 32bits
def ip_adress(ip):
    binary_ip = []

    for i in ip.split("."):
        convert = bin(int(i))[2:10]
        if len(convert) <= 8:
            add = ((8-len(convert))*"0")
            fullbin = (add+convert)
            binary_ip.append(fullbin)
        else:
            binary_ip.append(i)
    return binary_ip


#goes through ever value in subnet and ip adress, return the multipled of both lists (network adress calculation) 
def networkadress(ip, sub):
    net_adr_temp = ""
    net_adr = []
    ip_list = []
    sub_list = []
    # loops through each in the ip adress and append to comma seperated list
    for i in ip:
        for val_1 in i:
            ip_list.append(val_1)
            
    for y in sub:
        for val_2 in y:
            sub_list.append(val_2)
    #combine 2 list into 1 lists
    temp = list(zip(ip_list,sub_list))
    # multiple each value
    for i in temp:
        net_adr_temp = net_adr_temp + str(int(i[0])*int(i[1]))

    startnum = 0
    endnum = 8
    count = 0
    while count < 4:
        net_adr.append(net_adr_temp[startnum:endnum])
        startnum += 8
        endnum += 8
        count += 1
    
    return net_adr


#calculates the prefix value based 32bit ips
def prefix(net_pre):
    host_bits = 32 - net_pre
    num_hosts = (2** host_bits)-2
    sub_net_bi_temp = (net_pre*"1")+(host_bits*"0")
    sub_net_bi = []
    startnum = 0
    endnum = 8
    count = 0
    while count < 4:
        sub_net_bi.append(sub_net_bi_temp[startnum:endnum])
        startnum += 8
        endnum += 8
        count += 1

    return [sub_net_bi, net_pre, host_bits, num_hosts]

 
# take binary network adress converts to broadcast
def subnet_adress(ip_bin,network = int,host = int):
    
    ip_temp = ""

    for i in ip_bin:
        for y in i:
            ip_temp = (ip_temp) + y

    broadcast_temp = (ip_temp[0:network]) + ("1"*host)
    startip_temp = (ip_temp[0:31]) + ("1")
    lastip_temp = (ip_temp[0:network]) + ("1"*(host-1)) + ("0")

    broadcast = []
    startip = []
    lastip = []
    startnum = 0
    endnum = 8
    count = 0
    while count < 4:
        broadcast.append(broadcast_temp[startnum:endnum])
        startip.append(startip_temp[startnum:endnum])
        lastip.append(lastip_temp[startnum:endnum])
        startnum += 8
        endnum += 8
        count += 1

    return [broadcast,startip,lastip]


# calculate sub networks
def sub_to_prefix(pre):
    net = 0
    host = 0
    for i in pre:
        net += i.count("1")
        host += i.count("0")
    

    power_hosts = int(2**host)
    host_all = power_hosts
    nets = 1
    count = 0
    max_use = []
    while count < host:
        print(f'Futher Subnets: {nets} Usable hosts: {host_all-2}')
        if count == (host-2):
            max_use.append(nets)
            max_use.append(host_all-2)
        
        nets = nets * 2
        host_all = int(host_all/2)
        count += 1


    print(f"Maximum usable Subnets: {max_use[0]} With {max_use[1]} usable hosts each")
    num_hosts = (power_hosts)-2
    
    return net, host, num_hosts

def sub_convert_prefix(pre):
    net = 0
    host = 0
    for i in pre:
        net += i.count("1")
        host += i.count("0")
    return net, host


# prints binary
def convert_ip():
    ip = input('input ip adress ex: 192 or 192.225.0.168')
    temp = ip_adress(ip)
    for i in temp:
        print(i, end='.')

#converts from ip adress binary to decimal
def convert_binary():
    net_adr = str(input("binary adress:")).split(".")
    print(f"ipadress decimal: {int(net_adr[0],2)}.{int(net_adr[1],2)}.{int(net_adr[2],2)}.{int(net_adr[3],2)}")

    
def networkadress_print():
    ip = str(input('input ip adress format 192.168.0.2'))
    sub = str(input('input subnet format 255.255.0.0'))
    sub_bin = ip_adress(sub)
    prefix_val = sub_convert_prefix(sub_bin)
    all_print(ip,prefix_val[0])


def hosts_print():
    ip = input('input ip adress this format 192.168.0.2')
    net_pre = int(input('input prefix ex: 24'))
    all_print(ip,net_pre)


def all_print(ip,net_pre=int):
    #ip = input('input ip adress this format 192.168.0.2')
    #net_pre = int(input('input value of host prefix ex: 24'))
    ip_bin = ip_adress(ip)
    sub_bin = prefix(net_pre)
    net_val = networkadress(ip_bin, sub_bin[0])
    subnet_adr = subnet_adress(net_val, sub_bin[1], sub_bin[2])

    print(f"ip adress in binary: {ip_bin[0]}.{ip_bin[1]}.{ip_bin[2]}.{ip_bin[3]}")
    print(f"ip adress: {ip}")
    print("-"*100)
    print(f"subnet adress binary: {sub_bin[0][0]}.{sub_bin[0][1]}.{sub_bin[0][2]}.{sub_bin[0][3]}")
    print(f"subnet adress: {int(sub_bin[0][0],2)}.{int(sub_bin[0][1],2)}.{int(sub_bin[0][2],2)}.{int(sub_bin[0][3],2)}")
    print("-"*100)    
    print(f"network adr binary: {net_val[0]}.{net_val[1]}.{net_val[2]}.{net_val[3]}")
    print(f"network ip dec: {int(net_val[0],2)}.{int(net_val[1],2)}.{int(net_val[2],2)}.{int(net_val[3],2)}")
    print("-"*100)
    print(f"First ip binary: {subnet_adr[1][0]}.{subnet_adr[1][1]}.{subnet_adr[1][2]}.{subnet_adr[1][3]}")
    print(f"First ip dec: {int(subnet_adr[1][0],2)}.{int(subnet_adr[1][1],2)}.{int(subnet_adr[1][2],2)}.{int(subnet_adr[1][3],2)}")
    print("-"*100)
    print(f"Last ip binary: {subnet_adr[2][0]}.{subnet_adr[2][1]}.{subnet_adr[2][2]}.{subnet_adr[2][3]}")
    print(f"Last ip dec: {int(subnet_adr[2][0],2)}.{int(subnet_adr[2][1],2)}.{int(subnet_adr[2][2],2)}.{int(subnet_adr[2][3],2)}")
    print("-"*100)
    print(f"Broadcast ip binary: {subnet_adr[0][0]}.{subnet_adr[0][1]}.{subnet_adr[0][2]}.{subnet_adr[0][3]}")
    print(f"Broadcast ip dec: {int(subnet_adr[0][0],2)}.{int(subnet_adr[0][1],2)}.{int(subnet_adr[0][2],2)}.{int(subnet_adr[0][3],2)}")
    print("-"*100)
    print(f"Prefix: {sub_bin[1]} host bits: {sub_bin[2]} possible hosts: {sub_bin[3]}")
    print("-"*100)
    print("All nets created:")
    sub_to_prefix(sub_bin[0])

  
def calculator():
    print('to convert a decimal ip adress to binary input "1"')
    print('to convert a binary ip adress to decimal input "2"')
    print('to find info givin 192.168.24.2/24 input "3"')
    print('to find info given Ip 192.168.24.2and subnet mask 255.0.0.0 input "4"')
    print("-"*100)

    val = int(input('input number to choose calculator'))
    if val == 1:
        convert_ip()
    elif val == 2:
        convert_binary()
    elif val == 3:
        hosts_print()
    elif val == 4:
        networkadress_print()
    else:
        print("input a number 1-4")
    

calculator()

#192.168.200.139
#255.255.255.224
#27
```
